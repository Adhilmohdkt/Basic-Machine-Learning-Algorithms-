```python
fetures_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>=1]
```

```python
print(fetures_with_na)
```

```python
cat_nullfetures=[cat for cat in fetures_with_na if df[cat].dtype==object]
```

```python
print(cat_nullfetures)
```

Filling MIssing Values

```python
for i in fetures_with_na :
    if df[i].dtype==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].median())
```

```python
df.isnull().sum()
```

```python
df.drop('CustomerID',inplace=True,axis=1)
```

```python
cat_features=[features for features in df.columns if df[features].dtype=='O']
print ('categorical features are :',cat_features)
```

```python
num_features=[features for features in df.columns if df[features].dtype!='O']
print('numrical feattures are = ',num_features)
if 'ProdTaken' in num_features:
    num_features.remove('ProdTaken')
print('numrical feattures are = ',num_features)    
```

```python
x=df.drop(['ProdTaken'],axis=1)
y=df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
one=OneHotEncoder(drop='first')
std=StandardScaler()
colm=ColumnTransformer(
    [
        ('OneHot',one,cat_features),
        ('scaling',std,num_features)
    ]
)
```

```python
colm
```

```python
x_train=colm.fit_transform(x_train)
```

```python
pd.DataFrame(x_train)
```

```python
x_test=colm.transform(x_test)
pd.DataFrame(x_test)
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,recall_score,precision_score
```

```python
models={
    'RandomForest':RandomForestClassifier(),
    'Decision Tree':DecisionTreeClassifier(),
    'Logistic Regression':LogisticRegression()
}
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    y_pred_train=model.predict(x_train)
    #model testing on train data set  
    acc_train=accuracy_score(y_train,y_pred_train)
    class_train=classification_report(y_train,y_pred_train)
    #model testing on test data set
    acc_test=accuracy_score(y_test,y_pred)
    class_test=classification_report(y_test,y_pred)

    print(list(models.keys())[i])
    print('--MODEL PERFOMANCE ON TRAIN DATA --\n')
    print('Train accuracy = ',acc_train)
    print('Train classification report  ',class_train)
    print('--MODEL PERFOMANCE ON TEST DATA --\n')
    print('Test accuracy = ',acc_test)
    print('Train classification report  ',class_test)
    print("Finished:", list(models.keys())[i])


       
    
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_params = {"max_depth": [5, 8, 15, None, 10],
             "max_features": [5, 7, "auto", 8],
             "min_samples_split": [2, 8, 15, 20],
             "n_estimators": [100, 200, 500, 1000]}
```

```python
randcv_models=[
    ("Random_forest",RandomForestClassifier(),rf_params)
]
```

```python
model_param = {}
for name,selected,param in randcv_models:

    rand=RandomizedSearchCV(estimator=selected,param_distributions=param,verbose=2
                            ,n_jobs=-1,
                            cv=5,
                            n_iter=100)
    rand.fit(x_train,y_train)
    model_param[name] = rand.best_params_
    
```

```python
for i in model_param:
    print('---BEST PERAMETERS FOR MODEL',i,'ARE---')
    print(model_param[i])
```

```python
models={
    'RandomForest':RandomForestClassifier(n_estimators= 1000, min_samples_split= 2, max_features= 8, max_depth = None),
}
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    y_pred_train=model.predict(x_train)
    #model testing on train data set  
    acc_train=accuracy_score(y_train,y_pred_train)
    class_train=classification_report(y_train,y_pred_train)
    #model testing on test data set
    acc_test=accuracy_score(y_test,y_pred)
    class_test=classification_report(y_test,y_pred)

    print(list(models.keys())[i])
    print('--MODEL PERFOMANCE ON TRAIN DATA --\n')
    print('Train accuracy = ',acc_train)
    print('Train classification report  ',class_train)
    print('--MODEL PERFOMANCE ON TEST DATA --\n')
    print('Test accuracy = ',acc_test)
    print('Train classification report  ',class_test)
    
```
