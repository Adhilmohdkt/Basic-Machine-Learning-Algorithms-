```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Data/Travel.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'].replace({'Fe Male':'Female'},inplace=True)
```

```python
df['MaritalStatus'].value_counts()
```

```python
df['MaritalStatus'].replace({'Single':'Unmarried'},inplace=True)
```

```python
fetures_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>=1]
```

```python
cat_nullfetures=[cat for cat in fetures_with_na if df[cat].dtype==object]
```

```python
for i in fetures_with_na :
    if df[i].dtype==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].median())
```

```python
df.isnull().sum()
```

```python
df.drop('CustomerID',inplace=True,axis=1)
```

```python
cat_features=[features for features in df.columns if df[features].dtype=='O']
print ('categorical features are :',cat_features)
```

```python
num_features=[features for features in df.columns if df[features].dtype!='O']
print('numrical feattures are = ',num_features)
if 'ProdTaken' in num_features:
    num_features.remove('ProdTaken')
print('numrical feattures are = ',num_features) 
```

```python
x=df.drop(['ProdTaken'],axis=1)
y=df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
one=OneHotEncoder(drop='first')
std=StandardScaler()
colm=ColumnTransformer(
    [
        ('OneHot',one,cat_features),
        ('scaling',std,num_features)
    ]
)
```

```python
x_train=colm.fit_transform(x_train)
```

```python
x_test=colm.transform(x_test)
pd.DataFrame(x_test)
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,recall_score,precision_score
```

```python
models={
    "Adaboost":AdaBoostClassifier(),
    "Decision tree":DecisionTreeClassifier(),
    "Logisic Regression":LogisticRegression()

}
for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred=model.predict(x_test)
    acc_train=accuracy_score(y_train,y_pred_train)
    class_train=classification_report(y_train,y_pred_train)
    #model testing on test data set
    acc_test=accuracy_score(y_test,y_pred)
    class_test=classification_report(y_test,y_pred)

    print(list(models.keys())[i])
    print('--MODEL PERFOMANCE ON TRAIN DATA --\n')
    print('Train accuracy = ',acc_train)
    print('Train classification report  ',class_train)
    print('--MODEL PERFOMANCE ON TEST DATA --\n')
    print('Test accuracy = ',acc_test)
    print('Train classification report  ',class_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
ada_params={
    "n_estimators":[50,60,70,80,90],
    "algorithm":['SAMME','SAMME.R']
}
```

```python
sel_models=[
    ("ADA",AdaBoostClassifier(),ada_params)
]
```

```python
mod_peram={}
for name,slected,param in sel_models:
    rand=RandomizedSearchCV(estimator=slected,
                            param_distributions=param,
                                n_jobs=-1,
                                n_iter=70,
                                cv=5,
                                verbose=2)
    rand.fit(x_train,y_train)
    mod_peram[name]=rand.best_params_
    
```

```python
for i in range(len(mod_peram)):
    print("Best perameters are",list(mod_peram.values())[i])
```

```python
models={
    "Adaboost":AdaBoostClassifier(n_estimators= 70, algorithm= 'SAMME')

}
for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred=model.predict(x_test)
    acc_train=accuracy_score(y_train,y_pred_train)
    class_train=classification_report(y_train,y_pred_train)
    #model testing on test data set
    acc_test=accuracy_score(y_test,y_pred)
    class_test=classification_report(y_test,y_pred)

    print(list(models.keys())[i])
    print('--MODEL PERFOMANCE ON TRAIN DATA --\n')
    print('Train accuracy = ',acc_train)
    print('Train classification report  ',class_train)
    print('--MODEL PERFOMANCE ON TEST DATA --\n')
    print('Test accuracy = ',acc_test)
    print('Train classification report  ',class_test)
```
