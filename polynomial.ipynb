```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
''
```

```python
X = 6 * np.random.rand(100, 1) - 3
y =0.5 * X**2 + 1.5*X + 2 + np.random.randn(100, 1)
```

```python
plt.title('Data distribution')
sns.scatterplot(x=X.ravel(),y=y.ravel(),color='green')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
Linear=LinearRegression(n_jobs=-1)
Linear.fit(x_train,y_train)
```

```python
y_pred=Linear.predict(x_test)
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
print('r2 Score of the linear model is :',score)
```

```python
plt.figure(figsize=(5,5))
plt.title('Linear Regression Model')
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,Linear.predict(x_train))
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial=PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=polynomial.fit_transform(x_train)
x_test_poly=polynomial.transform(x_test)
```

```python
Linear.fit(x_train_poly,y_train)
y_pred_poly=Linear.predict(x_test_poly)
```

```python

from sklearn.metrics import r2_score
score=r2_score(y_pred_poly,y_test)
print('r2 Score of the Polynomial model is :',score)
```

```python

plt.suptitle('Polynomial Model',color='Black')
plt.scatter(x_train,Linear.predict(x_train_poly))
plt.scatter(x_train,y_train)
```
