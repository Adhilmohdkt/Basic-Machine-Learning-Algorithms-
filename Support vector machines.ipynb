```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=10000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_teain,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
sns.scatterplot(x=x_train[:,0],y=x_train[:,1],hue=y_teain)
```

```python
from sklearn.svm import SVC
model=SVC()
tuning={
    'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
    'C':[0.1,0.2,0.001,1000,10000,100]

}
```

```python
model.fit(x_train,y_teain)
y_pre=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
class1=classification_report(y_pre,y_test)
print(class1)
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=tuning,n_jobs=-1,cv=5)
```

```python
grid.fit(x_train,y_teain)
gpred=grid.predict(x_test)
```

```python
class2=classification_report(gpred,y_test)
print(class2)
```
