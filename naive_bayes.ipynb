```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
data.info()
```

```python
data
```

```python
data['day'].unique()
data['time'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
le=data['time']
y=label.fit_transform(le)
```

```python
dropped=data.drop(columns=['time'])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
preprocessing=ColumnTransformer(
    transformers=[(
        'cat1',OneHotEncoder(),['sex','smoker','day'],
)], remainder='passthrough')
processed=preprocessing.fit_transform(dropped)
```

```python
col_names=preprocessing.get_feature_names_out()
x=pd.DataFrame(processed,columns=col_names)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
acc=accuracy_score(y_pred,y_test)
clas1=classification_report(y_pred,y_test)
print('accuracy =',acc)
print('classification Report = ',clas1)
```
